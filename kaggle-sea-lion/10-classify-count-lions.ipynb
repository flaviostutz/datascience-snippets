### Mask image so only desired lion colors remain

```python
img_file = INPUT_DIR + 'Train/0.jpg'
im0 = cv2.imread(img_file)
im = im0[2000:3000,3000:4000,:]
utils.show_image(im, is_bgr=True)
```

### Find candidate areas by Selective Search

```python
import skimage
regions = skimage.segmentation.felzenszwalb(im, scale=200, sigma=0, min_size=10)
utils.show_image(regions, size=6)
```

```python
print('selectivesearch')
img_lbl, regions = selectivesearch.selective_search(im, scale=200, sigma=0, min_size=10)

print('filtering')
regions2 = []
for a in regions:
    size = a['size']
    r = a['rect']
    regions2 = regions2 + [r]
```

```python
print('showing')
im2 = im.copy()
print(len(regions2))
count = 0
for r in regions2:
    x, y, w, h = r
    if(50<w<150 and 50<h<150 and 0.7<(w/h)<1.3):
        cv2.rectangle(im2, (x, y), (x+w, y+h), (0,255,0), 2)
        count += 1
print(count)
utils.show_image(im2, is_bgr=True, size=14)
```
