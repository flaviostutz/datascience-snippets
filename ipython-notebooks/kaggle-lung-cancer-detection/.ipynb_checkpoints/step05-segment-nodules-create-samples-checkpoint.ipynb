```python
import sys
import h5py
import numpy as np
import shutil
import os
import datetime
import logging
import random

from modules.logging import logger
import modules.logging
import modules.utils as utils
from modules.utils import Timer
import modules.logging
```

```python
def lookup_nodules_simple(lung_pixels):
    id = random.randint(0,999999)
    utils.show_slices(lung_pixels, "lung" + str(id), nr_slices=12, cols=4, size=7)
    
    
    
    utils.show_slices(lung_pixels, "nodules" + str(id), nr_slices=6, cols=3, size=5)
    
```

```python
def generate_nodule_samples(input_x_ds, input_y_ds, output_x_ds, output_y_ds):
    output_index = 0
    for i in range(len(input_x_ds)):
        ts = Timer('Looking for nodules on patient lung ' + str(i))
        nodules_pixels = lookup_nodules_simple(input_x_ds[i])
        nodules_label = input_y_ds[i]
        for nodule_pixels in nodules_pixels:
            output_x_ds[output_index] = nodule_pixels
            output_y_ds[output_index] = nodules_label
            output_index = output_index + 1
        ts.stop()
        break #REMOVE LATER
    return output_index
```

```python
def start_processing(input_dir, input_image_dims, input_name, output_image_dims, output_dir):
    logger.info('Creating node samples. input_dir='+ input_dir + ' output_dir=' + output_dir)
    
    logger.info('Preparing output dir')
    utils.mkdirs(output_dir, dirs=['images'], recreate=True)

    modules.logging.setup_file_logger(output_dir + 'out.log')

    t = Timer('Creating nodule samples')
    len_input_x_ds = None
    input_dataset_file = dataset_path(input_dir, input_name, input_image_dims)
    with h5py.File(input_dataset_file, 'r') as input_h5f:
        logger.info('input x shape={}'.format(input_h5f['X'].shape))
        input_x_ds = input_h5f['X']
        input_y_ds = input_h5f['Y']
        len_input_x_ds = len(input_x_ds)
        len_input_y_ds = len(input_y_ds)
        
        output_h5f, output_x_ds, output_y_ds = utils.create_xy_datasets(output_dir, "nodules", output_image_dims, 9999):
        with output_h5f:
            total_nodules = generate_nodule_samples(input_x_ds, input_y_ds, output_x_ds, output_y_ds)
            #FIXME: RESIZE H5 DATASET TO total_nodules
        
        utils.validate_dataset(output_dir, "nodules", output_image_dims, save_dir=output_dir + 'images/')
    t.stop()
```
